```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.drop(['PassengerId','Name','Ticket','Cabin','Embarked', 'Fare'],axis='columns',inplace=True)
```

```python
y_test = test[['Pclass','Sex','Age','SibSp','Parch']]
```

```python
y_test.head()
```

```python
y_test.Sex = test.Sex.map({'male': 1, 'female': 2})
```

```python
y_test.Age = test.Age.fillna(test.Age.mean())
```

```python
inputs = train.drop('Survived',axis='columns')
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
target = train.Survived
target.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(inputs, target)
```

```python
inputs
```

```python
y_predicted = model.predict(y_test)
y_predicted
```
