```python
#Carregamento do log do gps - creator="Garmin Connect"
parsedXML = et.parse( "2890908.gpx" )
root = parsedXML.getroot()
namespace = '{http://www.topografix.com/GPX/1/0}'
idx=1
root.tag
```

```python
coordlst = []
elelst = []
timelst =[]
for child in root[idx]:
    for ch0 in child.findall("{http://www.topografix.com/GPX/1/0}trkpt"):
        coordlst.append(ch0.attrib)        
        for ch1 in ch0.findall("{http://www.topografix.com/GPX/1/0}time"):        
            timelst.append(ch1.text)

df_tmp_lat = pd.DataFrame(coordlst,columns=['lat','lon'])
crd = zip(iter(df_tmp_lat['lat']),iter(df_tmp_lat['lon']))
zip_trilha = zip(crd,crd)
df_trilha = pd.DataFrame(list(zip_trilha),columns=['origem','destino']).reset_index(drop=True)


df_tmp_lat['tempo'] = timelst
t = iter(df_tmp_lat['tempo'])
zip_duracao = zip(t,t)

df_tmp_tempo = pd.DataFrame(list(zip_duracao),columns=['saida','chegada'])
df_trilha['saida'] = df_tmp_tempo['saida'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%dT%H:%M:%S'))
df_trilha['chegada'] = df_tmp_tempo['chegada'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%dT%H:%M:%S'))

df_trilha
```

```python
#calculo da distancia
geolocator = Nominatim(user_agent="geoprocess", timeout=20)


torigem = list(df_trilha['origem'])
tdestino = list(df_trilha['destino'])
lstdist = []
trk = zip(torigem,tdestino)
for o,d in trk:
    lstdist.append(geodesic(o,d))

df_trilha['distancia'] = lstdist
df_trilha.head()

```

```python
#calculo da duração
tsaida = list(df_trilha['saida'])    
tchegada = list(df_trilha['chegada'])
lstduracao = []
durac = zip(tchegada,tsaida)
for chegada, saida in durac:
    lstduracao.append(chegada-saida)

df_trilha['duracao'] = lstduracao
df_trilha
```
